# Building df

```python
import pandas as pd
import numpy as np
```

```python
df_ligands1 = pd.read_csv('ligands_fingerprints.csv',sep=';')
df_ligands1
```

```python
df_ligands2= pd.read_csv('ligands_misc_info.csv',sep=';')
df_ligands2
```

```python
df_ligands = pd.DataFrame()
df_ligands['ID'] = df_ligands2['name']
df_ligands['Smiles'] = df_ligands2['smiles']
df_ligands['Stoichiometry'] = df_ligands2['stoichiometry']
df_ligands['Charge'] = df_ligands1['charge']
df_ligands['Nb atoms'] = df_ligands1['n_atoms']
df_ligands
```

```python
pd.DataFrame.to_csv(df_ligands, 'Separate ligands info.csv')
```
